```python
import requests
from bs4 import BeautifulSoup
link = "https://fr.wikipedia.org/wiki/Berb%C3%A8res"
page = requests.get(link)
soup = BeautifulSoup(page.content, 'html.parser')
article_data = {}
headings = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
paragraphs = soup.find_all('p')
current_heading = None
for item in headings + paragraphs:
    if item.name.startswith('h'):
        current_heading = item.get_text()
        article_data[current_heading] = []
    elif current_heading:
        article_data[current_heading].append(item.get_text())
for heading, paragraph_list in article_data.items():
    print(heading)
    for paragraph in paragraph_list:
        print("- ", paragraph)
    print()
```
